```python
def Normalize(Dataset):
    StandardDeviation = Dataset.std()
    for i in range(len(Dataset.columns.tolist())):
        Names = Dataset.columns.tolist()
        Name = Names[i]
        if Name != 'ID':
            print(f"The name of the column that we are working on now is: \'{Name}\'")
            EditedColumn = Dataset[Name].copy()
            print('The mean is: ', EditedColumn.mean())
            print('The Standard Deviation is: ', StandardDeviation[i])
            print('Z Score Normalizing...')
            EditedColumn = (( EditedColumn - EditedColumn.mean() )/ StandardDeviation[i])
            print('Updating Dataset...')
            Dataset[Name] = EditedColumn
            print('Done!')
    return Dataset
```

```python
def UniqueinDataset(LatestDataset):
    uniques = {}
    columns = LatestDataset.columns.tolist()
    for i in range(len(columns)):
        name = columns[i]
        amount = len(LatestDataset[name].unique())
        uniques[name] = amount
    return uniques

```
